## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

subtotal = df["qty"] * df["unit_price"]

df["subtotal"] = subtotal.round(2)
df.head()




```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

total_weight = df["qty"] * df["unit_weight"]
df["total_weight"] = total_weight.round(2)

#shipping_price = if df["total_weight"] <= 50: return df["total_weight"] * 10 else: return df["total_weight"] * 7

def shipping_price(total_weight):
    if total_weight <= 50:
        return total_weight * 10
    return total_weight * 7

# "Apply" the function to the amount column
df['shipping_price'] = df['total_weight'].apply(shipping_price).round(2)
df.head()


```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

no_tax_total_price = df["subtotal"] + df["shipping_price"]
sales_tax = no_tax_total_price * .0925
total_price = no_tax_total_price + sales_tax

df["total_price"] = total_price.round(2)
df.head()

```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

subtotal_cost = df["qty"] * df["unit_cost"]
total_cost = subtotal_cost + df["shipping_price"]

df["cost"] = total_cost.round(2)
df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price

profit = df["total_price"] - df["cost"]

df["profit"] = profit.round(2)
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

#top_client_spend = df.loc[df["client_id"] == client_list]



def client_function(client_id, column):
    client_df = df.loc[df["client_id"] == client_id, column]
    return round(client_df.sum(), 2)

for client in client_list:
    print(f"{client} : ${client_function(client, 'total_price')}")

```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

dataframe_dict = {"client_id": client_list}

output_variable = ["qty", "shipping_price", "total_price", "cost", "profit"]

for variable in output_variable:
    dataframe_dict[variable] = [client_function(client_id, variable) for client_id in client_list]

summary_df = pd.DataFrame(dataframe_dict)
ascend_df = summary_df.sort_values("profit", ascending= False)
ascend_df

#client_id, qty, shipping_price, total_price, cost, profit
```
